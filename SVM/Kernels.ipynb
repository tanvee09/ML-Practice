```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from mpl_toolkits.mplot3d import Axes3D
from sklearn import svm
```

```python
X, Y = make_circles(n_samples=500, noise=0.02)
print(X.shape, Y.shape)
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.show()
```

```python
svc = svm.SVC() # default kernel is RBF
```

```python
svc.fit(X, Y)
```

```python
print(svc.score(X, Y) * 100)
```

```python
svc2 = svm.SVC(kernel='linear')
svc2.fit(X, Y)
print(svc2.score(X, Y) * 100)
```

```python
svc2 = svm.SVC(kernel='poly')
svc2.fit(X, Y)
print(svc2.score(X, Y) * 100)
```

```python
svc2 = svm.SVC(kernel='sigmoid')
svc2.fit(X, Y)
print(svc2.score(X, Y) * 100)
```

### Custom Kernel

```python
def custom_kernel(x1, x2) :
    return np.square(np.dot(x1, x2.T))
```

```python
svc2 = svm.SVC(kernel=custom_kernel)
svc2.fit(X, Y)
print(svc2.score(X, Y) * 100)
```
