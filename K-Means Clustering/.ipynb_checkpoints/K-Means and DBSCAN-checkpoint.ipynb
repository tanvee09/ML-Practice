```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles, make_moons
```

```python
X, Y = make_moons(n_samples=200, noise=0.1)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = Y)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
```

```python
km.fit(X)
```

```python
centers = km.cluster_centers_
label = km.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c = label)
plt.scatter(centers[:, 0], centers[:, 1], color = "magenta", marker='x')
plt.show()
```

# DBSCAN
## (Density Based Spatial Clustering of Applications with Noise)
- Based upon the idea that a cluster is a high density area surrounded by low density region
- Starts by exploring a small area, if density is "good enough" it is considered as a part of cluster, and explores neighbours and so on
- So it groups regions based on density

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=0.1, min_samples=5)
```

```python
dbs.fit(X)
```

```python
ypred = dbs.fit_predict(X)
```

```python
print(ypred)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = ypred)
plt.show()
```

```python
dbs = DBSCAN(eps=0.3, min_samples=6)
ypred = dbs.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c = ypred)
plt.show()
```

```python
X, Y = make_moons(n_samples=200, noise=0.1)
plt.scatter(X[:, 0], X[:, 1], c = Y)
plt.show()
```

```python
dbs = DBSCAN(eps=0.22, min_samples=6)
ypred = dbs.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c = ypred)
plt.show()
```
