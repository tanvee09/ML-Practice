```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
collage = np.zeros((430, 430, 3), dtype = 'uint8')
```

```python
bl = cv2.imread('bottom_left.jpg')
bottom_left = cv2.cvtColor(bl, cv2.COLOR_BGR2RGB)
bottom_left = cv2.resize(bottom_left, (200, 200))
collage[220:420, 10:210, :] = bottom_left
```

```python
br = cv2.imread('bottom_right.jpg')
bottom_right = cv2.cvtColor(br, cv2.COLOR_BGR2RGB)
bottom_right = cv2.resize(bottom_right, (200, 200))
collage[220:420, 220:420, :] = bottom_right
```

```python
tl = cv2.imread('top_left.jpg')
top_left = cv2.cvtColor(tl, cv2.COLOR_BGR2RGB)
top_left = cv2.resize(top_left, (200, 200))
collage[10:210, 10:210, :] = top_left
```

```python
tr = cv2.imread('top_right.jpg')
top_right = cv2.cvtColor(tr, cv2.COLOR_BGR2RGB)
top_right = cv2.resize(top_right, (200, 200))
collage[10:210, 220:420, :] = top_right
```

```python
c = cv2.imread('center.jpeg')
center = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)
center = cv2.resize(center, (100, 100))
center_img = np.zeros((120, 120, 3))
center_img[10:110, 10:110, :] = center
collage[155:275, 155:275, :] = center_img
```

```python
plt.imshow(collage)
plt.axis("off")
plt.show()
```
