# Multivariate Normal Distribution

Example:   

Normal Distribution with 2 parameters  
Dataset of a monkey having
- Weight of each monkey
- Height of each monkey

+ve correlation b/w weight and height

```python
# u = [ux, uy, ....] {Column vectors and not row vectors}
# spread = [sigma(x), sigma(y), ....]

# covariance matrix (sigma) = [[sigma(xx), sigma(xy)],
#                              [sigma(xy), sigma(yy)]]
# sigma(xx) -> spread along x-axis(1st dimension) -> sigma(x)
# sigma(xy) -> how x and y are correlated (=0 => change in x doesn't make any change in y, eg. height and color)

# X = [X1, X2, ..., Xm] (m features)
# u = [u1, u2, ..., um] These are column vectors

# P(x) = 
```

![Multivariate Normal Distribution](https://i.stack.imgur.com/EOyQI.png)

```python
mean = np.array([0.0, 0.0])
cov = np.array([[1, 0], [0, 1]])

dist = np.random.multivariate_normal(mean, cov, 500)
print(dist.shape)
print(dist)
```

```python
plt.scatter(dist[:, 0], dist[:, 1], alpha = 0.5)
plt.show()
```

```python
mean2 = np.array([0.0, 0.0])
cov2 = np.array([[1, 0.5], [0.5, 1]])

dist2 = np.random.multivariate_normal(mean2, cov2, 500)
print(dist2.shape)
```

```python
plt.scatter(dist2[:, 0], dist2[:, 1], alpha = 0.5)
plt.show()
```

```python
mean3 = np.array([0.0, 0.0])
cov3 = np.array([[1, 0.8], [0.8, 1]])

dist3 = np.random.multivariate_normal(mean3, cov3, 500)
print(dist3.shape)
```

```python
plt.scatter(dist3[:, 0], dist3[:, 1], alpha = 0.5)
plt.show()
```

```python
mean4 = np.array([5.0, 6.0])
cov4 = np.array([[1.3, 0.2], [0.2, 3]])

dist4 = np.random.multivariate_normal(mean4, cov4, 500)
print(dist4.shape)
```

```python
plt.scatter(dist4[:, 0], dist4[:, 1], alpha = 0.5)
plt.show()
```

```python
plt.scatter(dist3[:, 0], dist3[:, 1])
plt.scatter(dist4[:, 0], dist4[:, 1])
plt.show()

# sigma(x) is more in blue so spread is more
```
