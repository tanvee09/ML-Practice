```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('bulbasaur.png')
```

```python
print(img.shape) # BGR channels because OpenCV interprets as BGR instead of RGB
```

```python
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.imshow(rgb_image)
plt.axis("off")
plt.show()
```

```python
# In openCV, rows & columns are also switched. So, it is (Column, Row, Channel)
crop_image = rgb_image[100:700, 200:600, :]
plt.imshow(crop_image)
plt.show()
```

```python
crop_image.shape
```

```python
total_pixels = crop_image.shape[0] * crop_image.shape[1]
print(total_pixels)
total_integers_needed = total_pixels * crop_image.shape[2]
print(total_integers_needed)
```

```python
cropped_image = cv2.resize(crop_image, (100, 150))
print(cropped_image.shape)
```

```python
plt.imshow(cropped_image)
plt.axis("off")
plt.title("Bulbasaur")
plt.show()
```
