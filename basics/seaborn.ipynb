## Data Visualisation using Seaborn

```python
import seaborn as sns
import numpy as np
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic[0:5]
```

```python
tips = sns.load_dataset('tips')
```

```python
tips[0:5]
```

```python
tips.head()
```

## Bar Plot

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips)

# Mean aggregation
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.sum)
```

## Count Plot

```python
# Aggregation function is always count

sns.countplot(x = 'sex', data = tips)
```

```python
sns.countplot(x = 'time', data = tips)
```

```python
sns.countplot(x = 'smoker', data = tips)
```

```python
sns.countplot(x = 'day', data = tips)
```

## Box Plot

Quartiles  
Shows distribution of categorical data  
Bottom line is minm and top line is maxm  
Maxm lie in the box, middle line is mean  
The parts above are outliers

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

### Violin plot (same as boxplot)

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

## Distribution Plot

```python
sns.distplot(tips['total_bill'])
```

```python
# kde -> kernel density estimation
sns.distplot(tips['total_bill'], kde = False)
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 30)
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 100)
```

## KDE Graph

```python
sns.kdeplot(tips['total_bill'])
```

## Joint Plot

```python
# Right bar plot is tip, top bar plot is totat_bill

sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
# linear trend b/w tip and total_bill
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
```

## Pair Plot

```python
# Pairwise realtion among all features
sns.pairplot(data = tips)

# For same axes (diagonals), creates normal distribution of that variable
```

```python
sns.pairplot(data = tips, hue = 'sex')
```

```python
# Calculates correlation among non-categorical variables
tips.corr()
```

## Heat Map

Data should have variables on both rows and columns  
Example -> tips.corr(), not tips  
tips is a panda dataframe

```python
sns.heatmap(tips.corr())
# light color -> high correlation
# dark color -> low correlation
```

```python
sns.heatmap(tips.corr(), annot = True)
```

## Remove features that are highly correlated

### Pivot Table (to get data frames with both variable rows and variable columns)

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
# Panda function
# index -> rows
flights_pivot = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
flights_pivot
```

```python
sns.heatmap(flights_pivot)
```

```python
sns.heatmap(flights_pivot, cmap = "coolwarm")
```
