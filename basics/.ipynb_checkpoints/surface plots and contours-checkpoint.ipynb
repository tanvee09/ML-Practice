# Surface Plots | Data Visualisation
### Surface plots are used to
- Visualise Loss Functions in Machine Learning and Deep Learning
- Visualise State or State Value in Reinforcement Learning

```python
#from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = np.arange(3) + 1
B = np.arange(4) + 4

print(a)
print(b)
```

```python
a, b = np.meshgrid(A, B)
# a is repeated along the row b.shape[0] number of times
# b is repeated along the columns a.shape[0] number of times
# Dimsensions of a and b now = b.shape[0] x a.shape[0]
print(a)
print(b)
# For the surface (2-D), x is taken from a and y is taken from b
# Coordinates will be -> (1, 4), (2, 4), (3, 4)
#                        (1, 5), (2, 5), (3, 5) ...
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d') # in matplotlib's 3D api
plt.show()
# Gives axes in 3D
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a, b, a + b)
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a, b, a + b, cmap = 'coolwarm') # Red value is high, blue is low
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a, b, a + b, cmap = 'rainbow') # Red value is high, blue is low
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a, b, a**2 + b**2) # Red value is high, blue is low
plt.show()
```

```python
A = np.arange(-1, 1, 0.02)
B = A
a, b = np.meshgrid(A, B)
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a, b, a**2 + b**2, cmap = 'rainbow')
plt.show()
```

# Contour Plots

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.contour(a, b, a**2 + b**2, cmap = 'rainbow')
plt.title('Contour Plot')
plt.show()
```
