### Line Plot
y = mx + c

Requirements  
- Array of Xs
- Array of Ys  

Draw line passing through these points

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.style.use("dark_background") 
plt.plot(x, y1, color = "green", label = "Guava")
plt.plot(x, y2, color = "red", label = "Apple")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over time")
plt.show()
```

### Histogram

```python
np.random.randn?
```

```python
xsn = np.random.randn(5)
print(xsn)
```

```python
Xsn = np.random.randn(100)
```

```python
# Marks of students in an exam out of 100

sigma = 5 # To widen the curve (to 8 from 1)
u = 60 # To shift the centre rightwards (to 70 from 0)
X = np.round(Xsn*sigma + u)
print(X)
```

```python
x2 = np.round(Xsn * 5 + 40)
```

```python
plt.hist(X, alpha = 0.8, label = "Physics")
plt.hist(x2, alpha = 0.5, label = "Math") # alpha is opacity component
plt.ylabel("Prob/Freq Count of Students")
plt.xlabel("Marks Range")
plt.legend()
plt.show()
```
