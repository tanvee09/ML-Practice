### Gaussian/Normal Distribution

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
gnb = GaussianNB()
```

```python
X, Y = make_classification(n_samples = 200, n_features = 2, n_informative = 2, n_redundant = 0, random_state = 8)
```

```python
X[:5]
```

```python
Y[:5]
```

```python
plt.scatter(X[:, 0], X[:, 1], c = Y)
plt.show()
```

```python
print(X.shape, Y.shape)
```

```python
gnb.fit(X, Y)
```

```python
gnb.score(X, Y)
```

```python
ypred = gnb.predict(X)
ypred[:10]
```

```python
Y[:10]
```

```python
import numpy as np
acc = np.sum((ypred == Y)) / X.shape[0]
```
