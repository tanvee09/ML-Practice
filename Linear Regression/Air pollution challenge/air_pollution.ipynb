```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Train.csv')
```

```python
df.head()
```

```python
data = df.values
```

```python
X = data[:, :5]
y = data[:, 5]
```

```python
X = (X - np.mean(X, axis = 0)) / np.std(X, axis = 0)
```

```python
print(X.shape, y.shape)
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(X, y)
```

```python
for i in range(5) :
    plt.scatter(X[:, i], y)
    plt.show()
```

```python
x_test = pd.read_csv('Test.csv')
```

```python
x_test = x_test.values
```

```python
x_test
```

```python
y_ = model.predict(x_test)
```

```python
y_
```

```python
df_test = pd.DataFrame(y_, columns = ['target'])
```

```python
df_test.head()
```

```python
df_test.to_csv('predictions.csv', index_label = 'Id')
```
