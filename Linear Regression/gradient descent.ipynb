```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X - 5)**2
print(X)
print(Y)
```

**Goal** Given a function f(x), we want to find the value of x that minimizes f.

## Visualisation

```python
plt.style.use("seaborn")
plt.plot(X, Y)
plt.xlabel("X")
plt.ylabel("F(X)")
plt.show()
```

```python
x = 0
lr = 0.1 # Learning rate (Step size)
# 50 steps downhill

error = []

plt.plot(X, Y)

for i in range(50) :
    gradient = 2 * (x - 5)
    x = x - lr * gradient # lr * gradient is effective step size
    y = (x - 5) ** 2
    error.append(y)
    plt.scatter(x, y)
    print(x)
    
plt.show()
```

```python
plt.plot(error)
plt.show()
```

```python
x = 0
lr = 0.1 # Learning rate (Step size)
# 50 steps downhill

error = []

threshold = 10 ** (-3) # Ideally 10^(-3)

plt.plot(X, Y)

prev = x

for i in range(100) :
    gradient = 2 * (x - 5)
    x = x - lr * gradient # lr * gradient is effective step size
    y = (x - 5) ** 2
    error.append(y)
    plt.scatter(x, y)
    print(x)
    if abs(x - prev) < threshold :
        print("Stopped after " + str(i + 1) + " iterations")
        break
    prev = x
else :
    print("Stopped after 100 iterations")
    
plt.show()


plt.plot(error)
plt.show()
```
